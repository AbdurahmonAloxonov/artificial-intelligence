```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


```

```python
# Ma'lumotni yuklash
data = pd.read_csv('/content/athlete_training.csv')
X = data['Training Hours per Week'].values.reshape(-1, 1)
y = data['Running Time (minutes)'].values


```

```python
# Polynomial xususiyatlar (daraja = 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)
```

```python

# Modelni oâ€˜qitish
model = LinearRegression()
model.fit(X_poly, y)
```

```python

# Bashorat
y_pred = model.predict(X_poly)

```

```python

# Baholash
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

```

```python
# Grafik uchun tartib
sort_idx = X.flatten().argsort()
X_sorted = X[sort_idx].reshape(-1, 1)
X_poly_sorted = poly_features.transform(X_sorted)
y_curve = model.predict(X_poly_sorted)
```

```python

# Chizish
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_sorted, y_curve, color='red', label='Polynomial regression (degree=2)')
plt.title('Training Hours vs Running Time')
plt.xlabel('Training Hours per Week')
plt.ylabel('Running Time (minutes for 5 km)')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Natijalar
print("Polynomial regression results:")
print(f"MSE: {mse:.2f}")
print(f"RÂ²: {r2:.4f}")
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)

```

```python
# Bashorat funksiyasi
def predict_running_time(hours):
    hours_array = np.array([hours]).reshape(-1, 1)
    hours_poly = poly_features.transform(hours_array)
    predicted_time = model.predict(hours_poly)
    return predicted_time[0]
```

```python

# Namuna bashoratlar
print("\nSample predictions:")
test_hours = [1, 3, 5, 7, 10]
for h in test_hours:
    result = predict_running_time(h)
    print(f"If an athlete trains {h} hours/week, predicted running time: {result:.2f} minutes")
```
